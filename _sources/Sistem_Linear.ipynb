## penyelesaian sistem linier
penyelesaian sistem persamaan linier dg ,menggunakan elimininasi gaus

selesaikan persamaan berikut

$$
8x_1+x_2+x_3=19\\
x_1+2x_2+3x_3=9\\
2x_1+x_2+x_3=5
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Persamaan bidang: x1 - x2 + x3^2 x1 + x2 + 8x3^4 x1 + 2x2 - 3x3 = 3 + 18 - 2
coefficients = np.array([[1, -1, 1], [2, 1, 8], [4, 2, -3]])
constants = np.array([3, 18, -2])

# Persiapkan data titik-titik pada bidang
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Hitung z (nilai koordinat ketiga) menggunakan persamaan bidang
z = (constants[0] - coefficients[0,0]*x - coefficients[0,1]*y) / coefficients[0,2]

# Buat plot 3D menggunakan Matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot bidang
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100)

# Label sumbu
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
